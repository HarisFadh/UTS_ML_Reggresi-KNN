### Library

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('counter_strike_2.csv')
data
## memasukkan dan membaca file csv dengan library pada pandas.read / pd.read dan memberikknya nama data tersebut "data"
```

```python
data.describe()
```

```python
data.info()

## untuk mengetahu tipe data tiap kolom
```

### Mengubah Values Kolom yang bertipe Data String/Objek menjadi Numerik

```python
for col in ['map', 'bomb_planted', 'round_winner']:
    label_encoder = LabelEncoder()
    data[col] = label_encoder.fit_transform(data[col])

## dengan label encoder mengubah tipe data string kedalam numerik ada tiga data yang diubah yaitu 2 data tipe objek 
## dan 1 tipe boolean 
```

```python
data
```

### Memisahkan kolom target dari data 

```python
x = data.drop('round_winner', axis = 1)
y = data['round_winner']

# membagi antara data target(dependen) dan data independen
```

```python
x
```

```python
y
```

### Splitting Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

## melakukan pembagian antara data test dan data train yang saya bagi disini 0.2 atau 20%/80% data yang dibagi
```

### Mengetahui Dimensi dari data training 

```python
x_train.shape
```

### Modelling KNN

```python
modelKNN = KNeighborsClassifier(n_neighbors=5)
modelKNN.fit(x_train, y_train)

## Parameter ini menentukan jumlah tetangga terdekat yang akan digunakan oleh algoritma KNN untuk melakukan klasifikasi.
```

```python
hasil_prediksi_KNN = modelKNN.predict(x_test)
```

```python
print('Akurasi Model: ', accuracy_score(y_test, hasil_prediksi_KNN) * 100)

## didapatkan disini hasil test keakurasian dengan tingkat sekitar 72 ~~ 76 membuktikan bahwa kurang efektif menggunakan KNN
## dengan dataset menentukan siapa yang menang dalam round didalam game counter strike 2 yang diberikan
```
